```python
#seperating fraud transaction and non fraud transaction

def split_data(data):
    nonfraud=data[data['Class']==0]
    fraud=data[data['Class']==1]
    return nonfraud,fraud

nonfraud,fraud=split_data(data)
```

```python
fraud
```

```python
nonfraud
```

```python
fraud.shape[0]
```

```python
nonfraud.shape[0]
```

```python
nonfraud.describe()
```

```python
new_nonfraud=nonfraud.sample(n=473)  #sampling with 473 non fraud samples to match with 473 fraud samples
```

```python
combined=pd.concat([new_nonfraud,fraud],axis=0)
```

```python
combined
```

```python
combined['Class'].shape[0]
```

```python
combined['Class'].value_
```

```python
x=combined.drop('Class',axis=1)
y=combined['Class']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=5)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_predict=model.predict(x_train)
accuracy=accuracy_score(x_train_predict,y_train)
accuracy
```

```python
x_test_predict=model.predict(x_test)
accuracy=accuracy_score(x_test_predict,y_test)
accuracy
```
